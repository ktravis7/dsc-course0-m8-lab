## Data Cleaning

### Filtering aircrafts and events

We want to filter the dataset to include aircraft that the client is interested in an analysis of:
- inspect relevant columns

- filter the dataset

**STEP 3**

```python
#What i'll do next is to inspect the the relevant columns that the client is interested in an analysis of

#1. Aircraft categories, how many of them are there and any total number of missing values
print("Aircraft categories;") # print aircraft categories in the data
print(df["Aircraft.Category"].value_counts(dropna=False))

#2. Amateur builds. Client is only interested in professionally built pieces
print("\nAmateur built;")
print(df["Amateur.Built"].value_counts(dropna=False))

#3. Event date. From the df.info() this column is stored as an object and should be a datetime datatype
print("\nEvent date;")
print(df["Event.Date"].head())
```

**STEP 3.1. Data cleaning, filtering aircrafts and events**

```python
#After inspection weve noticed some few problems such as missing values and incorrect formats that we need to first fix
# In the aircraft category 56602 values are missing, we can drop these because client is specific on wanting airplanes only, rows without categories cant be classified
df = df.dropna(subset=["Aircraft.Category"])

# We will also drop the 102 NaN values for Amateur builds since we cant determine whether they were professional or amateur builds
df = df.dropna(subset=["Amateur.Built"])

# Finally, I will convert event date to a datetime format
df["Event.Date"] = pd.to_datetime(df["Event.Date"])
```

```python
# Now we filter the data based on clients needs
#1. Get the airplanes category only
df = df[df["Aircraft.Category"] == "Airplane"]

#2. Get the professional builds only. From the amateur builds method the professional ones are marked by boolean 'No'
df = df[df["Amateur.Built"] == "No"]

#3. Get only the events from the past 40 years
# So here we can get today's timestamp using the pd.timestamp special class and subtract it to the number of years we need (40) using DateOffset class
valid_date = pd.Timestamp.today() - pd.DateOffset(years=40)
df = df[df["Event.Date"] >= valid_date]

# Get the new shape
print("New filtered dataset shape", df.shape)
```

### Cleaning and constructing Key Measurables

Injuries and robustness to destruction are a key interest point for the client. Clean and impute relevant columns and then create derived fields that best quantifies what the client wishes to track. **Use commenting or markdown to explain any cleaning assumptions as well as any derived columns you create.**

**Construct metric for fatal/serious injuries**

*Hint:* Estimate the total number of passengers on each flight. The likelihood of serious / fatal injury can be estimated as a fraction from this.

**STEP 4**

```python
# First thing is to construct metrics for fatal/serious injuries using derived fields
# So inorder to estimate the total number of passengers we need to use the injuries figures and since it has the injuries numbers we can use that

passenger_injury_columns = [
    "Total.Fatal.Injuries",
    "Total.Serious.Injuries",
    "Total.Minor.Injuries",
    "Total.Uninjured"
]

df[passenger_injury_columns] = df[passenger_injury_columns].fillna(0) # fill in the missing values with 0 to make calculations smoother

# Now estimate the total number of passengers
df["Total.Passengers"] = (
    df["Total.Fatal.Injuries"] + df["Total.Serious.Injuries"] + df["Total.Minor.Injuries"] +df["Total.Uninjured"]
)
```

```python
# Total serious/fatal injuries
df["Fatal.Serious.Injuries"] = (
    df["Total.Fatal.Injuries"]
    + df["Total.Serious.Injuries"]
)


# Avoided dividing by zero
df["Total.Passengers"] = df["Total.Passengers"].replace(0, np.nan)

# Now get the probability of a passenger suffering a fatal or serious injury
df["Fatal.Serious.Rate"] = (
    df["Fatal.Serious.Injuries"]
    / df["Total.Passengers"]
)

print(df["Fatal.Serious.Rate"].describe())

print(
    "Records with a rate above zero:",
    (df["Fatal.Serious.Rate"] > 0).sum()
)

print(
    "Maximum injury rate:",
    df["Fatal.Serious.Rate"].max()
)

df.loc[
    df["Fatal.Serious.Rate"] > 0,
    [
        "Make",
        "Model",
        "Total.Passengers",
        "Total.Fatal.Injuries",
        "Total.Serious.Injuries",
        "Fatal.Serious.Rate"
    ]
].head(10)

#The reason for this is if we dont have the total number of passengers we cant really gage how fatal the accident was,
# for instance if we had an accident wuth 5 passengers and all 5 were seriously injured it was dangerous than having another with
# 50 and only 10 were injured, this is how we get a rating of seriousness of the accident
```

**Cleaning Other Columns Task**

1. Engine.Type
- Remove leading and trailing spaces
- Replace placeholder values such as "UNKNOWN" with missing values(np.nan)
- Standardize capitalization

2. Weather.Condition
- Standardize text formatting
- Remove leading and trailing spaces
- Replace placeholder values such as "UNK" with missing values

3. Number.of.Engines
- Remove impossible values
- Keep valid engine counts

4. Purpose.of.flight
- Group categories with very few observations into "Others" for ease in analysis

5. Broad.phase.of.flight
- Standardize capitalization
- Group rare phases into "Other"


```python
# Execution of cleaning tasks
# 1.Engine Type
df["Engine.Type"] = df["Engine.Type"].str.strip()
df["Engine.Type"] = df["Engine.Type"].replace("UNKNOWN", np.nan)
df["Engine.Type"] = df["Engine.Type"].str.upper()

# 2.Weather Condition
df["Weather.Condition"] = df["Weather.Condition"].str.upper()
df["Weather.Condition"] = df["Weather.Condition"].str.strip()
df["Weather.Condition"] = df["Weather.Condition"].replace("UNK", np.nan)

# 3.Number of Engines
df = df[df["Number.of.Engines"] > 0]

# 4.Purpose of Flight
purpose_counts = df["Purpose.of.flight"].value_counts()
rare_purpose = purpose_counts[purpose_counts < 50].index

df["Purpose.of.flight"] = df["Purpose.of.flight"].replace(
    rare_purpose,
    "Other"
)

#5. Broad Phase of Flight

phase_counts = df["Broad.phase.of.flight"].value_counts()
rare_phase = phase_counts[phase_counts < 50].index

df["Broad.phase.of.flight"] = df["Broad.phase.of.flight"].replace(
    rare_phase,
    "Other"
)
```
